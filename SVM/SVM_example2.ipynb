```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.datasets import make_blobs
```

```python
def plot_classifier(X, y, model):
    # Plot data
    plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap = 'cool')
    

    # Plot data with decision boundary
    ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    
    ## Create grid
    xx = np.linspace(xlim[0], xlim[1], 30)
    yy = np.linspace(ylim[0], ylim[1], 30)
    YY, XX = np.meshgrid(yy, xx)
    xy = np.vstack([XX.ravel(), YY.ravel()]).T
    
    Z = model.decision_function(xy).reshape(XX.shape)

    ## Plot decision boundary and margins
    ax.contour(XX, YY, Z, colors=['grey','red', 'grey'], levels=[-1, 0, 1], alpha=0.8,
               linestyles=['--', '-', '--'])

    ## Support vectors
    ax.scatter(model.support_vectors_[:, 0], model.support_vectors_[:, 1], s=100,
           linewidth=1, facecolors='none', edgecolors='grey')
    plt.show()


```

```python
# generate data
X,y = make_blobs(n_samples = 100, centers=2, random_state = 123, cluster_std = 3)
```

```python
X.shape
```

```python
plt.scatter(X[:,0],X[:,1], c=y, s=50, cmap='cool')
```

```python
# fit model 1
svc = svm.SVC(kernel = 'linear')
svc.fit(X,y)
```

```python
plot_classifier(X,y, svc)
```

```python
# fit model 1+
svc1 = svm.SVC(kernel = 'linear', C=0.1)
svc1.fit(X,y)
```

```python
plot_classifier(X,y, svc1)
```

```python
# fit model 2
svc_poly = svm.SVC(kernel = 'poly', C=0.1, degree = 3)
svc_poly.fit(X,y)
```

```python
plot_classifier(X,y, svc_poly)
```

```python
# fit model 3
svc_rbf = svm.SVC(kernel = 'rbf', C=0.1)
svc_rbf.fit(X,y)
```

```python
plot_classifier(X,y, svc_rbf)
```

```python
# fit model 3+
svc_rbf1 = svm.SVC(kernel = 'rbf', C=1)
svc_rbf1.fit(X,y)
```

```python
plot_classifier(X,y, svc_rbf1)
```
